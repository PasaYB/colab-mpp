```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca Data
data = pd.read_csv("/content/drive/My Drive/datacsv/muvi.csv", encoding='unicode_escape')

# Memilih kolom untuk sumbu x dan y
x = 'Year'
y = 'Movies Released'

# Mengelompokkan data dan menghitung total movies released per tahun untuk beberapa genre
genres = ['Action', 'Comedy', 'Drama', 'Thriller or Suspense', 'Adventure']

plt.figure(figsize=(10, 8))

# Plot line chart untuk setiap genre
for genre in genres:
    genre_data = data[data['Genre'] == genre]
    grouped_df = genre_data.groupby(x)[y].sum().reset_index()
    plt.plot(grouped_df[x], grouped_df[y], label=genre, marker='o', linestyle='-')

plt.title('Line Chart of Total Movies Released for Different Genres')
plt.xlabel(x)
plt.ylabel(f'Total {y}')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca Data
data = pd.read_csv("/content/drive/My Drive/datacsv/rice.csv", encoding='unicode_escape')

subset_data = data.head(20)

# Memilih hanya data untuk tahun 2020
data_2020 = data[data['Year'] == 2020]

# Memilih kolom untuk sumbu x dan y
x = 'Provinsi'
y = 'Production.(ton)'

# Menambahkan jarak antar bar
bar_width = 0.7  # Sesuaikan lebar bar sesuai kebutuhan
bar_positions = range(len(subset_data[x]))

# Fungsi untuk konversi label angka
def format_label(value):
    if value >= 1e6:
        return f'{value/1e6:.1f}Jt'
    elif value >= 1e3:
        return f'{value/1e3:.1f}Rb'
    else:
        return f'{int(value)}'

# Membuat bar chart vertikal dengan jarak antar bar
plt.figure(figsize=(12, 8))
bars = plt.bar(bar_positions, subset_data[y], color='orange', width=bar_width)
plt.title("Produksi Beras pada Tahun 2020")
plt.xlabel("Provinsi")
plt.ylabel("Produksi")
plt.xticks(bar_positions, subset_data[x], rotation=45, ha='right')  # Menyesuaikan rotasi dan posisi label sumbu x

# Menambahkan grid hanya pada sumbu x
plt.grid(axis='y')

# Menambahkan label di atas setiap bar dengan format khusus
for bar in bars:
    yval = bar.get_height()
    label = format_label(yval)
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, label, ha='center', va='bottom')

plt.gca().get_yaxis().get_major_formatter().set_scientific(False)  # Menonaktifkan notasi ilmiah pada sumbu y
plt.tight_layout()  # Menyesuaikan layout agar tidak tumpang tindih
plt.show()
```
